```python
print_uniquecol_values(df1)
```

```python
yes_no_col = ['Partner','Dependents', 'PhoneService','MultipleLines','OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport','StreamingTV',
              'StreamingMovies','PaperlessBilling','Churn']
```

```python
for col in yes_no_col:
    df1[col].replace({'Yes':1 , 'No':0}, inplace = True)
    
```

```python
for column in df1:
    print(f'{column}: {df1[column].unique()}')
    
```

```python
# for more than one categorical data, we use one-hot encoding
df2 = pd.get_dummies(data=df1 , columns = ['InternetService','Contract', 'PaymentMethod'], dtype = int) 
```

```python
df2
```

```python
df2.dtypes
```

```python
# In order to bring all the values ;ile the total charges and the monthly charges as well asthe tenure to a suitable range of 0 to 1 , we need to using a scaler function known as the min max scaler () function

cols_to_scale = ['tenure', 'MonthlyCharges', 'TotalCharges']
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df2[cols_to_scale] = scaler. fit_transform(df2[cols_to_scale])

```

```python
for column in df2:
    print(f'{column}: {df2[column].unique()}')
# this made our dataset ready for the training    
```

```python
x = df2.drop('gender', axis = 'columns')
x1 = x.drop('Churn', axis = 'columns')

y = df2['Churn']        
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(x1,y,test_size = 0.2, random_state = 5)
```

```python
len(X_train.columns)
```

```python
X_test.shape
```

```python
import tensorflow as tf
from tensorflow import keras

model = keras.Sequential([
    keras.layers.Dense(20, input_shape = (25,), activation='relu'),    # this defines how many neurons we want to have for the input layer, and here we will enter the same no. of neurons as the no of columns in our training dataset X
    
    keras.layers.Dense(1, activation='sigmoid'),
])

model.compile(optimizer = 'adam',
              loss = 'binary_crossentropy',       # we have our results in binary format, so we use the binary cross entropy
              metrics = ['accuracy'])

model.fit(X_train, Y_train, epochs = 1000)
              
    
```

```python
model.evaluate(X_test, Y_test)
```

```python
yp = model.predict(X_test)
yp[:5]          # this is the output for the x_test and it would return me a 2d array which we need to convert
```

```python
Y_test[:10]   # actual values of y
```

```python
y_pred = []
for elements in yp:
    if elements > 0.5:
        y_pred.append(1)
    else:
        y_pred.append(0)
```

```python
y_pred[:10]    # y predicted values by the model
```

```python
from sklearn.metrics import confusion_matrix , classification_report

print (classification_report(Y_test,y_pred))

```

```python
import seaborn as sn
cm = tf.math.confusion_matrix(labels=Y_test, predictions = y_pred)
plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel( 'Truth' )

```
